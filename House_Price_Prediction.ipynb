```python
#extracting numerical and categorical features
numeric_col = data.select_dtypes(include=['float64', 'int64'])
object_col = data.select_dtypes(include='object').columns
```

```python
#correlation matrix for numeric columns
correlation_matrix = numeric_col.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#transforming categoric columns into numeric ones
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in object_col:
    data[col] = label_encoder.fit_transform(data[col])
```

```python
#Splitting data
from sklearn.model_selection import train_test_split

X=data.drop(columns='price',axis=1)
y=data['price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
#importing the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#evaluating the model
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print ("Mean squared error : ", mse, end="\n" )
print ("R2 score : ", r2)
```
